````python
import bs4, re, urllib.request
acl2019html = open('all.html', 'r').read()
soup = bs4.BeautifulSoup(acl2019html, 'html.parser')
````

````python
def extract_papers(soup):
#    L = soup.findAll('a', attrs={'class':'align-middle', 'href': re.compile(r'aclweb\.org/')})
#     L = soup.findAll(
#         lambda tag:tag.name == "a" and 
#         tag.get("class", '') == "align-middle" and
#         re.search('', tag["href"])
#     )
    T = []
    ps = soup.select('p[class="d-sm-flex align-items-stretch"]')
    for p in ps:
        a = p.find('a', attrs={'class':'d-none'})
        title = p.find('strong').find('a')
#         title = p.findAll('span', class_= ["d-block"])
        T.append( (a['href'], " ".join(title.get_text().split())) ) 
    return T

def download_papers(T):
    for (url, title) in T:
        paper_ID =  re.match(r'https://www.aclweb.org/anthology/([W|P]\d+-\d+).pdf', url).group(1)
        filename = paper_ID + "_" + "_".join(title.split()) + '.pdf'
        print (filename)
        urllib.request .urlretrieve('https://www.aclweb.org/anthology/P19-1000', filename)
````

````python
T = extract_papers(soup)
download_papers(T)
````
